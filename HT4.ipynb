# 4. Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см


```python
M=174
std=8
X=182

Z=(X-M)/std
print(f'Z={round(Z,2)}')
```

Из таблицы накопленного нормального распределения следует, что за 1 стандартным отклонением лежит 0,8413 (84,13%) значений. А вероятность того, что СВ больше Х=182 будет равна 1-0,8413=0.1587

б). больше 190 см

```python
X=190

Z=(X-M)/std
print(f'Z={round(Z,2)}')
```

Из таблицы накопленного нормального распределения следует, за 2 стандартными отклонениями лежит 0,9772 (97,72%) значений. А вероятность того, что СВ больше Х=190 будет равна 1-0,9772=0,0228

в). от 166 см до 190 см

```python
from statistics import NormalDist

X1=166
X2=190

Z1=(X1-M)/std
Z2=(X2-M)/std
print(f'Z1={round(Z1,2)}\nZ2={round(Z2,2)}')
```

Из таблицы накопленного нормального распределения следует, что за -1 стандартным отклонением лежит 0,1587 (15,87%) значений, за 2 std -
  0,9772 А вероятность того, что СВ лежит от Z1 ло Z2 = P(Z2)-P(Z1)

```python
P=NormalDist().cdf(Z2)-NormalDist().cdf(Z1)
print(f'P={round(P,2)}')
```

г). от 166 см до 182 см

```python
X1=166
X2=182

Z1=(X1-M)/std
Z2=(X2-M)/std
print(f'Z1={round(Z1,2)}\nZ2={round(Z2,2)}')

P=NormalDist().cdf(Z2)-NormalDist().cdf(Z1)
print(f'P={round(P,2)}')
```

д). от 158 см до 190 см

```python
X1=158
X2=190

Z1=(X1-M)/std
Z2=(X2-M)/std
print(f'Z1={round(Z1,2)}\nZ2={round(Z2,2)}')

P=NormalDist().cdf(Z2)-NormalDist().cdf(Z1)
print(f'P={round(P,2)}')
```

е). не выше 150 см или не ниже 190 см

```python
X1=150
X2=190

Z1=(X1-M)/std
Z2=(X2-M)/std
print(f'Z1={round(Z1,2)}\nZ2={round(Z2,2)}')

P=NormalDist().cdf(Z1)+(1-NormalDist().cdf(Z2))
print(f'P={round(P,2)}')
```

ё). не выше 150 см или не ниже 198 см

```python
X1=150
X2=198

Z1=(X1-M)/std
Z2=(X2-M)/std
print(f'Z1={round(Z1,2)}\nZ2={round(Z2,2)}')

P=NormalDist().cdf(Z1)+(1-NormalDist().cdf(Z2))
print(f'P={round(P,3)}')
```

ж). ниже 166 см.

```python
X=166

Z=(X-M)/std
print(f'Z={round(Z,2)}')

P=NormalDist().cdf(Z)
print(f'P={round(P,3)}')
```

## 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
X=190
M=178
D=25

std=np.sqrt(D)
Z=(X-M)/std
print(f'Z={round(Z,2)}')
```

Коэффициент Z показывает на сколько сигм отклоняется рост человека=190 от матожидания
